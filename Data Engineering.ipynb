# Data Mounting

```python
from google.colab import drive
import pandas as pd
import os
from PIL import Image # For loading images

# --- 1. Mount Google Drive ---
try:
    drive.mount('/content/drive', force_remount=True) # force_remount can be useful
    print("Google Drive mounted.")
except Exception as e:
    print(f"Error mounting Google Drive: {e}")

# --- 2. Define Paths ---
# IMPORTANT: Adjust these paths to match your file locations in Google Drive.
# After mounting, 'My Drive' is typically the root of your personal drive.
base_drive_path = '/content/drive/My Drive/Proyek Akhir Pengsis/Dataset/tbx11k-simplified' # Common base path

# Path to your CSV file
csv_file_name = 'data.csv'
csv_file_path = os.path.join(base_drive_path, csv_file_name) # e.g., /content/drive/My Drive/data.csv
# If data.csv is in a subfolder, like 'MyProject/datasets/', change to:
# csv_file_path = os.path.join(base_drive_path, 'MyProject/datasets/', csv_file_name)


# Path to your folder containing images
image_folder_name = 'images' # The name of your image folder
image_folder_path = os.path.join(base_drive_path, image_folder_name) # e.g., /content/drive/My Drive/images/
# If images folder is in a subfolder, like 'MyProject/training_data/', change to:
# image_folder_path = os.path.join(base_drive_path, 'MyProject/training_data/', image_folder_name)

# --- 3. Load the CSV file ---
dataframe = None
if os.path.exists(csv_file_path):
    try:
        dataframe = pd.read_csv(csv_file_path)
        print(f"\nSuccessfully loaded '{csv_file_name}'.")
        # print("First 5 rows of the CSV data:")
        # print(dataframe.head())
    except Exception as e:
        print(f"Error loading CSV file '{csv_file_path}': {e}")
else:
    print(f"CSV file not found at: {csv_file_path}")
    print("Please check the 'csv_file_path' variable.")

# --- 4. Access and Load Images ---
image_files_list = []
loaded_image_example = None

if os.path.exists(image_folder_path) and os.path.isdir(image_folder_path):
    print(f"\nAccessing images in folder: {image_folder_path}")
    try:
        # List all files in the image folder
        all_files_in_folder = os.listdir(image_folder_path)
        # Filter for common image file extensions (you can add more)
        image_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff']
        image_files_list = [f for f in all_files_in_folder if os.path.splitext(f)[1].lower() in image_extensions]

        if image_files_list:
            print(f"Found {len(image_files_list)} image files. First few: {image_files_list[:5]}")

            # Example: Load the first image found
            first_image_name = image_files_list[0]
            first_image_path = os.path.join(image_folder_path, first_image_name)

            try:
                loaded_image_example = Image.open(first_image_path)
                print(f"Successfully loaded example image: '{first_image_name}' (format: {loaded_image_example.format}, size: {loaded_image_example.size})")
                # You can display the image in Colab if needed:
                # from IPython.display import display
                # display(loaded_image_example)
            except Exception as e:
                print(f"Error loading image '{first_image_path}': {e}")
        else:
            print(f"No image files found in '{image_folder_path}'.")
            print("Please check the folder contents and supported extensions.")

    except Exception as e:
        print(f"Error accessing image folder '{image_folder_path}': {e}")
else:
    print(f"\nImage folder not found or is not a directory: {image_folder_path}")
    print("Please check the 'image_folder_path' variable.")
```

```python
# This cell assumes the previous cell (mounting drive, loading CSV into 'dataframe',
# and populating 'image_files_list' and 'image_folder_path') has been run successfully.

from IPython.display import display # For displaying images and DataFrames nicely in Colab
from PIL import Image # For loading images
import os # For joining paths

# --- 1. Display first 20 rows of the DataFrame ---
print("--- Displaying CSV Data (First 20 Rows) ---")
if 'dataframe' in globals() and dataframe is not None:
    if not dataframe.empty:
        display(dataframe.head(20))
    else:
        print("The 'dataframe' is empty.")
else:
    print("The 'dataframe' variable does not exist or is None. Please ensure the previous cell ran correctly and loaded the CSV.")

# --- 2. Load and Display First 3 Images ---
print("\n--- Displaying First 3 Images ---")
if 'image_files_list' in globals() and 'image_folder_path' in globals() and image_files_list is not None and image_folder_path is not None:
    if image_files_list:
        print(f"Found {len(image_files_list)} image files in the list. Displaying up to the first 3:")
        images_to_display = 0
        for i, image_name in enumerate(image_files_list[:3]): # Iterate through the first 3 image names
            image_path = os.path.join(image_folder_path, image_name)
            print(f"\nProcessing image {i+1}: {image_name}")
            if os.path.exists(image_path):
                try:
                    img = Image.open(image_path)
                    print(f"Successfully loaded '{image_name}' (format: {img.format}, size: {img.size}, mode: {img.mode})")
                    display(img) # Display the image in Colab output
                    images_to_display += 1
                except Exception as e:
                    print(f"Error loading or displaying image '{image_path}': {e}")
            else:
                print(f"Image file not found at path: {image_path}. It might have been moved or deleted, or the path in 'image_files_list' is incorrect.")
        if images_to_display == 0 and len(image_files_list) > 0:
            print("No images were displayed. Check for errors above or if image paths are correct.")
        elif len(image_files_list) < 3 and images_to_display > 0 :
             print(f"Displayed {images_to_display} images (fewer than 3 available in the list or folder).")

    else:
        print("The 'image_files_list' is empty. No images to display.")
        print(f"This list should have been populated from the folder: {image_folder_path}")
else:
    print("One or more required variables ('image_files_list', 'image_folder_path') do not exist or are None.")
    print("Please ensure the previous cell ran correctly and populated these variables.")

print("\n--- Demonstration in this cell complete ---")
```

# Data Classification

```python
# This cell assumes 'dataframe' (loaded from your data.csv) and
# 'image_folder_path' (path to your images) are available from previous cells.

import pandas as pd
from PIL import Image # For opening images to check existence/validity
import os
import matplotlib.pyplot as plt # Kept for optional image display

# --- Configuration: Updated based on your data.csv inspection ---
IMAGE_FILENAME_COLUMN = 'fname'             # Corrected: Column in data.csv with image names
LABEL_COLUMN = 'target'                   # Corrected: Column in data.csv with 'tb' or 'no_tb'
POSITIVE_LABEL = 'tb'                     # 'tb' will be treated as the "positive" or "non-healthy" class
NEGATIVE_LABEL = 'no_tb'                  # 'no_tb' will be treated as the "negative" or "healthy" class

# Image dimensions for display consistency if you uncomment the display part
IMG_WIDTH = 128
IMG_HEIGHT = 128

# --- 1. Check if prerequisite variables exist ---
# These should have been defined and loaded in previous Colab cells.
if 'dataframe' not in globals() or dataframe is None:
    print("ERROR: 'dataframe' not found. Please ensure 'data.csv' was loaded into a DataFrame named 'dataframe' in a previous cell.")
    # For robustness, you might want to stop execution here if critical variables are missing.
    # For example: raise NameError("'dataframe' not found. Please load data.csv first.")
if 'image_folder_path' not in globals() or image_folder_path is None:
    print("ERROR: 'image_folder_path' not found. Please ensure this variable, pointing to your image directory, was defined in a previous cell.")
    # For example: raise NameError("'image_folder_path' not found.")

# Perform detailed checks only if dataframe exists to avoid further errors
if 'dataframe' in globals() and dataframe is not None:
    print(f"Using DataFrame with {len(dataframe)} entries.")
    print(f"Expected image filename column: '{IMAGE_FILENAME_COLUMN}' (Checking...)")
    if IMAGE_FILENAME_COLUMN not in dataframe.columns:
        print(f"  ERROR: Column '{IMAGE_FILENAME_COLUMN}' NOT found in DataFrame. Available columns: {dataframe.columns.tolist()}")
    else:
        print(f"  SUCCESS: Column '{IMAGE_FILENAME_COLUMN}' found in DataFrame.")

    print(f"Expected label column: '{LABEL_COLUMN}' (Checking...)")
    if LABEL_COLUMN not in dataframe.columns:
        print(f"  ERROR: Column '{LABEL_COLUMN}' NOT found in DataFrame. Available columns: {dataframe.columns.tolist()}")
    else:
        print(f"  SUCCESS: Column '{LABEL_COLUMN}' found in DataFrame.")
        # Further check for expected labels if column exists
        unique_labels_in_column = dataframe[LABEL_COLUMN].unique()
        print(f"    Unique labels found in '{LABEL_COLUMN}': {unique_labels_in_column}")
        if POSITIVE_LABEL not in unique_labels_in_column:
            print(f"    WARNING: Expected POSITIVE_LABEL '{POSITIVE_LABEL}' not found among unique labels.")
        if NEGATIVE_LABEL not in unique_labels_in_column:
            print(f"    WARNING: Expected NEGATIVE_LABEL '{NEGATIVE_LABEL}' not found among unique labels.")


# --- 2. Classify Images based on CSV Data (Direct Lookup) ---
print("\nStarting classification based on CSV data...")

# Proceed only if dataframe and necessary columns are confirmed to exist and are correctly configured
if 'dataframe' in globals() and dataframe is not None and \
   'image_folder_path' in globals() and image_folder_path is not None and \
   IMAGE_FILENAME_COLUMN in dataframe.columns and \
   LABEL_COLUMN in dataframe.columns:

    classifications_summary = []
    images_reported_count = 0
    images_not_found_count = 0
    unknown_labels_count = 0 # Counts labels in CSV not matching POSITIVE_LABEL or NEGATIVE_LABEL

    for index, row in dataframe.iterrows():
        try:
            img_filename = row[IMAGE_FILENAME_COLUMN]
            # The classification is directly from the CSV's label column
            classification_from_csv = row[LABEL_COLUMN]

            # Validate the label from CSV against expected labels
            if classification_from_csv not in [POSITIVE_LABEL, NEGATIVE_LABEL]:
                # This case should ideally not happen if POSITIVE_LABEL and NEGATIVE_LABEL encompass all relevant values in LABEL_COLUMN
                print(f"Info: Image '{img_filename}' has a label '{classification_from_csv}' which is not '{POSITIVE_LABEL}' or '{NEGATIVE_LABEL}'. It will be reported as is.")
                # You might decide to skip these or handle them differently based on your needs.
                # For now, we'll still report it if the file exists.
                # unknown_labels_count += 1 # Increment if you want to count these separately
                pass # Continue processing

            img_path = os.path.join(image_folder_path, img_filename)

            if not os.path.exists(img_path):
                print(f"Warning: Image file '{img_filename}' (Label in CSV: {classification_from_csv}) not found at {img_path}. Skipping.")
                images_not_found_count += 1
                continue

            # Optional: Verify if it's a loadable image
            try:
                with Image.open(img_path) as img_check:
                    pass # Successfully opened, it's likely a valid image format
            except Exception as img_err:
                print(f"Warning: File '{img_filename}' at {img_path} exists but could not be opened as an image: {img_err}. This might indicate a corrupted file or non-image format. Skipping.")
                images_not_found_count +=1 # Count as effectively "not found" for classification purposes
                continue

            # Report the classification
            print(f"Image: '{img_filename}' is classified as: '{classification_from_csv}' (based on CSV column '{LABEL_COLUMN}')")
            classifications_summary.append({'filename': img_filename, 'classification_from_csv': classification_from_csv})
            images_reported_count += 1

            # Optional: Display the first few images with their CSV classification
            # Ensure matplotlib.pyplot was imported as plt
            # if images_reported_count <= 3: # Display first 3 successfully processed images
            #     try:
            #         img_display = Image.open(img_path)
            #         if img_display.mode != 'RGB': # Ensure RGB for consistent display
            #             img_display = img_display.convert('RGB')
            #         img_display = img_display.resize((IMG_WIDTH, IMG_HEIGHT))
            #
            #         plt.figure(figsize=(4, 4))
            #         plt.imshow(img_display)
            #         plt.title(f"File: {img_filename}\nCSV Label ('{LABEL_COLUMN}'): {classification_from_csv}", fontsize=10)
            #         plt.axis('off')
            #         plt.show()
            #     except Exception as display_e:
            #         print(f"  Could not display image {img_filename}: {display_e}")

            if (index + 1) % 200 == 0: # Print progress update less frequently for large datasets
                print(f"  Processed {index + 1}/{len(dataframe)} CSV entries...")

        except KeyError as ke:
            # This error indicates a problem with accessing columns, should be caught by initial checks but good fallback.
            print(f"Error: Missing expected column in CSV for row {index}. Details: {ke}. Please ensure '{IMAGE_FILENAME_COLUMN}' and '{LABEL_COLUMN}' are correct and present in all rows.")
        except Exception as e:
            # General error handling for unexpected issues with a row.
            current_filename_in_row = row.get(IMAGE_FILENAME_COLUMN, 'Filename N/A due to error')
            print(f"An unexpected error occurred processing CSV row for image '{current_filename_in_row}': {e}")

    print("\n--- Classification Reporting Complete ---")
    print(f"Successfully reported classification for {images_reported_count} images based on CSV data.")
    if images_not_found_count > 0:
        print(f"Could not find or open {images_not_found_count} image files that were listed in the CSV.")
    if unknown_labels_count > 0: # This count relies on uncommenting the incrementer above
        print(f"Encountered {unknown_labels_count} entries with labels other than '{POSITIVE_LABEL}' or '{NEGATIVE_LABEL}'.")

else:
    print("\nSkipping classification process due to configuration errors or missing data.")
    if 'dataframe' not in globals() or dataframe is None:
        print("Reason: 'dataframe' is not available. Ensure data.csv is loaded.")
    elif 'image_folder_path' not in globals() or image_folder_path is None:
        print("Reason: 'image_folder_path' is not available. Ensure the path to your image directory is set.")
    elif IMAGE_FILENAME_COLUMN not in dataframe.columns :
        print(f"Reason: The filename column '{IMAGE_FILENAME_COLUMN}' is missing from the DataFrame.")
    elif LABEL_COLUMN not in dataframe.columns:
        print(f"Reason: The label column '{LABEL_COLUMN}' is missing from the DataFrame.")
    else:
        print("Reason: An unspecified configuration issue occurred. Please review setup.")


print("\n--- Image Classification (CSV Lookup - Corrected) Cell Complete ---")
```
